```
s = input("Enter 3 real number (sperated by space): ")
s = s.split()
a = float(s[0])
b = float(s[1])
c = float(s[2])
print(f"a = {a}, b={b}, c={c}")
import math

# Nhập giá trị a, b, c
a = float(input("Nhập hệ số a: "))
b = float(input("Nhập hệ số b: "))
c = float(input("Nhập hệ số c: "))

# Kiểm tra điều kiện
if a == 0:
    # Phương trình bậc nhất hoặc không hợp lệ
    if b == 0:
        print("Phương trình không hợp lệ (a = 0 và b = 0).")
    else:
        # Phương trình bậc nhất: bx + c = 0
        x = -c / b
        print(f"Phương trình bậc nhất có nghiệm: x = {x}")
else:
    # Phương trình bậc hai
    delta = b**2 - 4*a*c
    print(f"Delta = {delta}")

    if delta > 0:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print("Phương trình có hai nghiệm phân biệt:")
        print(f"x1 = {x1}")
        print(f"x2 = {x2}")
    elif delta == 0:
        x = -b / (2*a)
        print("Phương trình có nghiệm kép:")
        print(f"x1 = x2 = {x}")
    else:
        print("Phương trình vô nghiệm thực.")
```

```
# Nhập số nguyên dương n từ bàn phím
n = int(input("Nhập một số nguyên dương n: "))

# Khởi tạo biến tổng
tong = 0

# Dùng vòng lặp for để tính tổng các số từ 0 đến n
for i in range(n + 1):
    tong += i

# Hiển thị kết quả ra màn hình
print(f"Tổng các số từ 0 đến {n} là: {tong}")
```
